# 02 — Feature Engineering

Loads `rfm.parquet` and `timeseries_features.parquet` and
visualises the per-customer RFM distribution plus the time-series
lag/rolling features used by the forecasting model.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
plt.rcParams['figure.figsize'] = (10, 4)
rfm = pd.read_parquet('data/processed/rfm.parquet')
ts = pd.read_parquet('data/processed/timeseries_features.parquet')
print('rfm:', rfm.shape, '|', list(rfm.columns))
print('ts :', ts.shape, '|', list(ts.columns))
```

## RFM distributions

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, col in zip(axes, ['Recency', 'Frequency', 'Monetary']):
    s = rfm[col]
    if col == 'Monetary':
        s = s[s <= s.quantile(0.99)]
    ax.hist(s, bins=40, color='steelblue', edgecolor='white')
    ax.set_title(f'{col} (n={len(rfm):,})')
plt.tight_layout()
plt.show()
```

## Time-series feature columns

Lag/rolling features are built by `features/timeseries.py` from
the cleaned daily revenue series.

```python
lag_cols = [c for c in ts.columns if c.startswith('lag_') or c.startswith('roll_')]
print('lag/roll columns:', lag_cols)
if lag_cols:
    ax = ts[lag_cols[:3]].plot(subplots=True, figsize=(10, 6), linewidth=1.0)
    plt.tight_layout()
    plt.show()
```
